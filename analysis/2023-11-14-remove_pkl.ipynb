# Remove epoch time in processed data

```python
filepaths = []

root_dir = "/home/wmnlab/D/database"
# root_dir = "/Users/jackbedford/Desktop/MOXA/Code/data"
dates = [s for s in sorted(os.listdir(root_dir)) if os.path.isdir(os.path.join(root_dir, s))]
date_dirs = [os.path.join(root_dir, s) for s in dates]
# date_dirs = date_dirs[:-7]
# date_dirs

for date_dir in date_dirs:
    # print(date_dir)
    exp_dirs = [os.path.join(date_dir, s) for s in sorted(os.listdir(date_dir)) if os.path.isdir(os.path.join(date_dir, s)) and s != '.DS_Store']
    for exp_dir in exp_dirs:
        # print(exp_dir)
        dev_dirs = [os.path.join(exp_dir, s) for s in sorted(os.listdir(exp_dir)) if os.path.isdir(os.path.join(exp_dir, s)) and s != '.DS_Store']
        for dev_dir in dev_dirs:
            # print(dev_dir)
            trip_dirs = [os.path.join(dev_dir, s, 'data') for s in sorted(os.listdir(dev_dir)) if os.path.isdir(os.path.join(dev_dir, s)) and s != '.DS_Store']
            for trip_dir in trip_dirs:
                # print(trip_dir)
                try:
                    pcap_files = [os.path.join(trip_dir, s) for s in sorted(os.listdir(trip_dir)) if s.endswith('loss_latency.csv')]
                except:
                    continue
                # print(pkl_files)
                
                filepaths += pcap_files
                
pprint(filepaths)
```

```python
for filepath in filepaths:
    print(filepath)
    df = pd.read_csv(filepath)
    df[['seq', 'rpkg', 'frame_id']] = df[['seq', 'rpkg', 'frame_id']].astype('Int64')
    df[['seq', 'rpkg', 'frame_id', 'Timestamp', 'lost', 'excl', 'latency', 'xmit_time', 'arr_time']].to_csv(filepath, index=False)
    print(round((df['latency'] < 0).mean() * 100, 3), '%\t', round((df['latency'] > 0.1).mean() * 100, 3), '%')
```
