```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./downlink loss rate.csv', index_col=0)
df = df.T

box_1, box_2, box_3, box_4, box_5, box_6, box_7, box_8 = df['All'], df['B3'], df['B7'], df['B8'], df['B3B7'], df['B3B8'], df['B7B8'], df['LTE']
plt.figure()
labels = 'All', 'B3', 'B7', 'B8', 'B3B7', 'B3B8', 'B7B8', 'LTE'
plt.boxplot([box_1, box_2, box_3, box_4, box_5, box_6, box_7, box_8], labels=labels)
plt.title('Downlink Loss Rate')

plt.show()

###
df = pd.read_csv('./uplink loss rate.csv', index_col=0)
df = df.T

box_1, box_2, box_3, box_4, box_5, box_6, box_7, box_8 = df['All'], df['B3'], df['B7'], df['B8'], df['B3B7'], df['B3B8'], df['B7B8'], df['LTE']
plt.figure()
labels = 'All', 'B3', 'B7', 'B8', 'B3B7', 'B3B8', 'B7B8', 'LTE'
plt.boxplot([box_1, box_2, box_3, box_4, box_5, box_6, box_7, box_8], labels=labels)
plt.title('Uplink Loss Rate')

plt.show()
```
