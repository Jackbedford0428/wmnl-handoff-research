```python
import os
import sys
import shutil
from genericpath import exists
from pprint import pprint

target_dir = "/home/wmnlab/Desktop/dataset"
source_dir = "/home/wmnlab/D/database"
dates = ["2022-09-29", "2022-10-11"]
file_tags = [("data", "cimon", "_preproc.csv"),
             ("data", "diag_log", ".txt"),
             ("data", "diag_log", "_rrc.csv"),
             ("analysis", "diag_log_ho-info", ".csv"),
             ("raw", "server_pcap_UL", ".pcap"),
             ("raw", "server_pcap_DL", ".pcap"),
             ("raw", "client_pcap_BL", ".pcap"),
             ("analysis", "dwnlnk_udp_latency", ".csv"),
             ("analysis", "dwnlnk_udp_loss_timestamp", ".csv"),
             ("analysis", "uplnk_udp_latency", ".csv"),
             ("analysis", "uplnk_udp_loss_timestamp", ".csv"),
            ]

# --------------------- Util Functions ---------------------
def makedir(dirpath, mode=0):  # mode=1: show message, mode=0: hide message
    if os.path.isdir(dirpath):
        if mode:
            print("mkdir: cannot create directory '{}': directory has already existed.".format(dirpath))
        return
    ### recursively make directory
    _temp = []
    while not os.path.isdir(dirpath):
        _temp.append(dirpath)
        dirpath = os.path.dirname(dirpath)
    while _temp:
        dirpath = _temp.pop()
        print("mkdir", dirpath)
        os.mkdir(dirpath)

def savecopy(filepath, targetdir, filename=""):
    ### filename can be different from basename of filepath, can be used to rename a file.
    makedir(targetdir)
    if not filename:
        filename = os.path.basename(filepath)
    print("cp -p", filepath, os.path.join(targetdir, filename))
    shutil.copy2(filepath, os.path.join(targetdir, filename))  # reserve modified date and some meta data
```

### Start Copy

```python
def is_specified(filename):
    for item in file_tags:
        if filename.startswith(item[1]) and filename.endswith(item[2]):
            return True
    return False

for date in dates:
    makedir(os.path.join(target_dir, date))
    savecopy(os.path.join(source_dir, date, "experiment-settings.txt"), os.path.join(target_dir, date))
    _list1 = [item for item in os.listdir(os.path.join(source_dir, date)) if os.path.isdir(os.path.join(source_dir, date, item)) and item != "raw_data"]
    print(_list1)
    for dir1 in _list1:
        _list2 = [item for item in os.listdir(os.path.join(source_dir, date, dir1)) if os.path.isdir(os.path.join(source_dir, date, dir1, item))]
        print(_list2)
        for dir2 in _list2:
            _list3 = [item for item in os.listdir(os.path.join(source_dir, date, dir1, dir2)) if os.path.isdir(os.path.join(source_dir, date, dir1, dir2, item))]
            print(_list3)
            for dir3 in _list3:
                for dir4 in ["raw", "data", "analysis"]:
                    _list_files = os.listdir(os.path.join(source_dir, date, dir1, dir2, dir3, dir4))
                    for filename in _list_files:
                        if is_specified(filename):
                            savecopy(os.path.join(source_dir, date, dir1, dir2, dir3, dir4, filename), os.path.join(target_dir, date, dir1, dir2, dir3))
```

### Calculate folder's size

```python
# assign size
size = 0

# get size
for path, dirs, files in os.walk(target_dir):
    for f in files:
        fp = os.path.join(path, f)
        size += os.path.getsize(fp)
        print(fp)

# display size
if size > 1 << 30:
    print("Folder size: {:.2f} GiB".format(size/(1 << 30)))
elif size > 1 << 20:
    print("Folder size: {:.2f} MiB".format(size/(1 << 20)))
elif size > 1 << 10:
    print("Folder size: {:.2f} KiB".format(size/(1 << 10)))
else:
    print("Folder size: {} bytes".format(size))
```
