```python
import os, sys
import shutil

def makedir(dirpath, mode=0):  # mode=1: show message, mode=0: hide message
    if os.path.isdir(dirpath):
        if mode:
            print("mkdir: cannot create directory '{}': directory has already existed.".format(dirpath))
        return
    ### recursively make directory
    _temp = []
    while not os.path.isdir(dirpath):
        _temp.append(dirpath)
        dirpath = os.path.dirname(dirpath)
    while _temp:
        dirpath = _temp.pop()
        print("mkdir", dirpath)
        os.mkdir(dirpath)
```

```python
targetdir = os.path.join(database, 'temp')
for date in dates:
    db1 = os.path.join(targetdir, date)
    makedir(db1)
    for (expr, (times, traces)), devices in zip(exps.items(), _devices):
        db2 = os.path.join(db1, expr)
        makedir(db2)
        for dev in devices:
            db3 = os.path.join(db2, dev)
            makedir(db3)
            for trace in traces:
                db4 = os.path.join(db3, trace)
                makedir(db4)
                for folder in folders:
                    shutil.copytree(os.path.join(database, date, expr, dev, trace, folder), os.path.join(db4, folder)) 
```
