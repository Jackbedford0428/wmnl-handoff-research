```python
import os
from os.path import isdir, isfile

def detect_event(filepath, tags):
    fp = open(filepath, 'r')
    lines = fp.readlines()  # neglect '\n' when reading the file
    for i, line in enumerate(lines):
        for tag in tags:
            if tag in line:
                print(f'{filepath}: line {i}')
                print(line)

def dfs(visited, node, filetags, tags, level=0):  # function for dfs
    if node in visited:
        return
    visited.add(node)
    ### TODO: 'node' is an absolute pathname (string) for a file or a directory.
    if os.path.basename(node).startswith(filetags):
        detect_event(node, tags)
    ### TODO END
    if isfile(node):
        return
    children = [os.path.join(node, s) for s in sorted(os.listdir(node))]
    for child in children:
        dfs(visited, child, filetags, tags, level+1)
    ### print message in the end
    if level == 0:
        print("**********************************************************************")
```

```python
def dfs(visited, node, filetags1, filetags2, tags, level=0):  # function for dfs
    if node in visited:
        return
    visited.add(node)
    ### TODO: 'node' is an absolute pathname (string) for a file or a directory.
    if os.path.basename(node).startswith(filetags1) and os.path.basename(node).endswith(filetags2):
        detect_event(node, tags)
    ### TODO END
    if isfile(node):
        return
    children = [os.path.join(node, s) for s in sorted(os.listdir(node))]
    for child in children:
        dfs(visited, child, filetags1, filetags2, tags, level+1)
    ### print message in the end
    if level == 0:
        print("**********************************************************************")
```
