```python
import os
from os.path import isdir, isfile

def dfs(visited, node, level=0):  # function for dfs
    if node in visited:
        return
    visited.add(node)
    ### print file structure
    for i in range(level):
        if i == level-1:
            print("|___", end='')
        else:
            print("|   ", end='')
    if level == 0 or isdir(node):
        print(node)
    else:
        print(os.path.basename(node))
    ### TODO: 'node' is an absolute pathname (string) for a file or a directory.
    ### TODO END
    if isfile(node):
        return
    children = [os.path.join(node, s) for s in sorted(os.listdir(node))]
    for child in children:
        dfs(visited, child, level+1)
    ### print message in the end
    if level == 0:
        print("**********************************************************************")
```

```python
root = "/home/wmnlab/D/database"
visited = set()  # Set to keep track of visited nodes of graph (tree).
dfs(visited, root)
```
